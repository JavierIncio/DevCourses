```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r-")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
# Object Oriented Method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left axis, bottom axis, width, heigth

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0,0.0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.5,0.4])

axes1.plot(x,y)
axes1.set_title("LARGER PLOT")

axes2.plot(y,x,"r-")
axes2.set_title("SMALLER PLOT")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].set_title("Second Plot")

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("my_picture.png", dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Squared")
ax.plot(x,x**3, label="X Cubed")

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y,color="purple", linewidth=1, linestyle="-", marker='*', markersize=20,
#        markerfacecolor="yellow",markeredgewidth=2, markeredgecolor="green")
ax.plot(x,y,color="#FF8C00", lw=2, ls="--") # RGB Hex Code

# Zoom
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
