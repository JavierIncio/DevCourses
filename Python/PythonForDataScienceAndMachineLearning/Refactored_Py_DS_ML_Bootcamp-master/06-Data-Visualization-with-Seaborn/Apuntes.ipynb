```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips["total_bill"],bins=40)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.pairplot(data=tips, hue="sex", palette="coolwarm")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
sns.barplot(x="sex",y="total_bill",data=tips, estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",dodge=True)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
sns.swarmplot(x="day",y="total_bill",data=tips, color="black")
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="violin")
```

```python
flights = sns.load_dataset("flights")
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr(numeric_only=True)
tc
```

```python
sns.heatmap(tc,annot=True,cmap="coolwarm")
```

```python
fp = flights.pivot_table(index="month", columns="year", values="passengers", observed=True)
fp
```

```python
sns.heatmap(fp,cmap="coolwarm",linecolor="black",lw=3)
```

```python
sns.clustermap(fp,cmap="coolwarm",standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```
